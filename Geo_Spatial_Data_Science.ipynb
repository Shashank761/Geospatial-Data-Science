# **Reading the data from the ShapeFile**

Hence this is geodataframe so use geopandas libray to read the file instead of simple pandas

```
shp_data_1 = gpd.read_file('TM_WORLD_BORDERS-0.3.shp')
```

# **Plotting the whole GeoDataFrame**


```
fig, ax = plt.subplots(figsize = (10,10))
shp_data_1.plot(ax=ax)
plt.show()
```

As the it has the point for all the country of the world , so it show a world map as its plot

# **Plotting the graph for record having specific feature**

Here we specify the country name column to be **'INDIA'**

```
fig, ax = plt.subplots(figsize = (10,10))
shp_data_1[shp_data_1['NAME']=='India'].plot(ax=ax)
plt.show()
```

We don't have the record or says data for indiviual city or place of a country , so we can only plot a counry image with the help of any of feature such as ** Population in 2005 , longitude , lalitide , iso code ** and many more.

# **Use of longitude and laltitude feature**

Points on the centroid of the country available in the dataset based on the logitude and latitude given

```
px.scatter_geo(shp_data_1 , lat="LAT", lon="LON")
```

**As we hover over any country it will sow its longitude and latitude**

# **A moving plot with covers all the country having featues given below**

```
fig = px.scatter_mapbox(
    shp_data_1, 
    lat="LAT", 
    lon="LON", 
    color = "NAME", 
    size="POP2005", 
    color_continuous_scale=px.colors.cyclical.IceFire, 
    animation_frame= "NAME",
    size_max=15, 
    zoom=1)
fig.show()
```

Larger the circle on coutry means greater the population of that country in 2005
